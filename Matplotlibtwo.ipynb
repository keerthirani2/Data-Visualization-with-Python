```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import folium
df = pd.read_csv("canadian_immegration_data.csv")
df.sample(20)
```

```python
A = df.sort_values(['Total'], ascending = False, axis=0, inplace = True)
A
```

```python
years = list(map(str, range(1980,2014)))
df_top5 = df.head()
df_top5 = df_top5[years].transpose()
B = df_top5
B
```

```python
B.plot(kind='area')
plt.title('Top5')
plt.ylabel("no.of immigratnts")
plt.xlabel('Years')
plt.show()
```

```python
#histograms
df['2013'].plot(kind ='hist',color='red')
plt.title("Histograms of the 195 country")
plt.ylabel("No of countries")
plt.xlabel("No of immigration")
plt.show()
```

```python
#for the better understanding the histogrms we use the numpy
count, bin_edges = np.histogram(df['2013'])
df['2013'].plot(kind='hist', xticks = bin_edges)
plt.show()
```

```python
#piechart
Y = df.groupby('Continent', axis =0).sum()
Y
```

```python
#marking on the map
canada_map = folium.Map(location=[56.130,-106.35],zoom_start=4)

ontario = folium.map.FeatureGroup()
#styleing the feature group
ontario.add_child(
    folium.features.CircleMarker(
    [51.25,-85.32],radius=5, 
     color='red',fill_color='Red'))
#add the feature grop to the map
canada_map.add_child(ontario)
#label the marker
folium.Marker([51.25,-85.32],
              popup='Ontario').add_to(canada_map)
canada_map
      
```
